```python
Django app

Django application consists of project and app, it also generates an automatic base directory for the app, 
so we can focus on writing code (business logic) rather than creating app directories.

The difference between a project and app is, a project is a collection of configuration files and apps 
whereas the app is a web application which is written to perform business logic.
```

```python
Steps for Creating application in Django.

1) First create the project using command "django-admin startproject my_project".
2) Then set the path inn your command prompt by using "cd my_project".
3) Create the project by using the command "django-admin startapp my_application".

Note : in creating project and app i use my_project and my_application these are not default you can use any name.

4) In our application there are some default python files. they are __init__.py,admin.py,apps.py,models.py,tests.py,views.py.
5) Go to views.py follow above views.py code and save it.
6) Then come back in project in project also there is some default python files. 
7) They are __init__.py,asgi.py,settings.py,urls.py,wsgi.py.
8) Go to urls.py and follow urls.py code and save it.
9) Then come to command prompt use command "python manage.py runserver". if not execute use "python manage.py migrate" it will
    remove all migrations and then execute by again "python manage.py runserver". go to web browser paste "http://127.0.0.1:8000/".
10) In that link give our function name which we used in views.py and urls.py like "http://127.0.0.1:8000/hello". output will comes.
11) Follow below output.
```
